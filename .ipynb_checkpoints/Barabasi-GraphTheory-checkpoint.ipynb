```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.barabasi_albert_graph(10,2)
```

```python
nx.draw(G)
```

```python
G = nx.barabasi_albert_graph(100000,2)
```

```python
Degrees = np.array(nx.degree(G))[:,1]
```

```python
vals, freq = np.unique(Degrees, return_counts=True)
```

```python
plt.loglog(vals,freq,'k.')
X = np.array([1.0, 1000.0])
plt.loglog(X,500000*X**(-3),'r--')
plt.show()
```

```python
for r in range(10):
    G = nx.barabasi_albert_graph(100000,2)
    Degrees = np.array(nx.degree(G))[:,1]
    vals, freq = np.unique(Degrees, return_counts=True)
    plt.loglog(vals,freq,'.')
X = np.array([1.0, 1000.0])
plt.loglog(X,500000*X**(-3),'r--')
plt.ylabel(r'$f(k)$', size = 20)
plt.xlabel(r'$k$', size = 20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```
