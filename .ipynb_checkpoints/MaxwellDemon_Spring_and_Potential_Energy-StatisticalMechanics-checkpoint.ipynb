```python
import random
import numpy as np
import matplotlib.pyplot as plt

m     = 1       #se definen variables del sistema
k     = 1
gamma = 1
b     = 1
lim   = 2       # limite de la variable estocastica
dt = 0.001

y = [0]         # condiciones iniciales
t = [0]
vy = 0
Fy = 0
relative = 0
tout = 0
num_steps = 3   # numero de veces que va a subir
num_steps2 = 30 # numero de veces que va a iterar con una misma variable estocastica

for step2 in range(num_steps2):
    ys = random.uniform(-lim, lim)
    yevolve = y[step2-1] + vy*dt
    y.append(yevolve) 
    Fy      = (-k*(relative-ys))-(m*gamma*vy)
    vy      = vy + Fy*dt/m
    tout    = tout + 1
    t.append(tout)
    if y[step2] > relative:
        relative = 2*y[step2]

plt.xlabel("Time")
plt.ylabel("Height")
plt.title("Height of the Particle")
plt.plot(t,y)
```
