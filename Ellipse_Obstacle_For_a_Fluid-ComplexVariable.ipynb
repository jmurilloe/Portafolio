```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
lim = 11
eppr = 26
x,y = np.meshgrid(np.linspace(-lim,lim,eppr),np.linspace(-lim,lim,eppr))
a = 10
b = 5
c = np.sqrt(a*2 - b*2)
vx=0*x*y
vy=0*x*y

num_steps = eppr
for step1 in range(num_steps):
    for step2 in range(num_steps):
        if x[step1,step2] > 0:
            vx[step1,step2] = (1/2)*(c**2 + (a + b)**2 + (c**2 - (a + b)**2)*((1/2)*x[step1,step2] + (1/4)*(4*x[step1,step2]*y[step1,step2]**2 + 2*x[step1,step2]*(-c**2 + x[step1,step2]**2 - y[step1,step2]**2))/np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2))/np.sqrt(-1/2*c**2 + (1/2)*x[step1,step2]**2 - 1/2*y[step1,step2]**2 + (1/2)*np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2)))/c**2
            vy[step1,step2] = (1/2)*(c**2 - (a + b)**2)*(-1/2*y[step1,step2] + (1/4)*(4*x[step1,step2]**2*y[step1,step2] - 2*y[step1,step2]*(-c**2 + x[step1,step2]**2 - y[step1,step2]**2))/np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2))/(c**2*np.sqrt(-1/2*c**2 + (1/2)*x[step1,step2]**2 - 1/2*y[step1,step2]**2 + (1/2)*np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2)))
        elif x[step1,step2] < 0:
            vx[step1,step2] = (1/2)*(c**2 + (a + b)**2 - (c**2 - (a + b)**2)*((1/2)*x[step1,step2] + (1/4)*(4*x[step1,step2]*y[step1,step2]**2 + 2*x[step1,step2]*(-c**2 + x[step1,step2]**2 - y[step1,step2]**2))/np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2))/np.sqrt(-1/2*c**2 + (1/2)*x[step1,step2]**2 - 1/2*y[step1,step2]**2 + (1/2)*np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2)))/c**2
            vy[step1,step2] = -1/2*(c**2 - (a + b)**2)*(-1/2*y[step1,step2] + (1/4)*(4*x[step1,step2]**2*y[step1,step2] - 2*y[step1,step2]*(-c**2 + x[step1,step2]**2 - y[step1,step2]**2))/np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2))/(c**2*np.sqrt(-1/2*c**2 + (1/2)*x[step1,step2]**2 - 1/2*y[step1,step2]**2 + (1/2)*np.sqrt(4*x[step1,step2]**2*y[step1,step2]**2 + (-c**2 + x[step1,step2]**2 - y[step1,step2]**2)**2)))
N = np.sqrt(vx**2+vy**2)
i = (vx)/(N)
j = (vy)/(N)
figure, axes = plt.subplots()
Drawing_colored_circle = patches.Ellipse((0,0), 2*a, 2*b, color='black')

axes.set_aspect( 1 )
axes.add_artist( Drawing_colored_circle )
plt.title( 'Ellipse obstacle' )
plt.quiver(x,y,i,j)
plt.show()
```
