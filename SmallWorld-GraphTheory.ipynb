```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
Nv = 1000      # Nodes
ki = 10        # Initial degree of the nodes
p  = np.logspace(-4.0,0,20)
C        = []  # Start parameters
errorC   = []
L        = []
errorL   = []
partialC = []
partialL = []
cmax =  0
lmax =  0
ecmax = 0
elmax = 0

for pi in p:
    
    for i in range(20):
        
        G = nx.watts_strogatz_graph(Nv, ki, pi)
        ctt = nx.average_clustering(G)
        partialC.append(ctt)
        ltt = nx.average_shortest_path_length(G)
        partialL.append(ltt)
        
    ct = np.average(partialC)
    C.append(ct)                           #  Add 20 networks with probability pi average
    ect = np.std(partialC)
    errorC.append(ect)
    lt = np.average(partialL)
    L.append(lt) 
    elt = np.std(partialL)
    errorL.append(elt)
    
    partialC = []                          # Clean memory
    partialL = []
    
    if ct >= cmax:
        cmax = ct
    if lt >= lmax:
        lmax = lt
    
ecmax = errorC[C.index(cmax)]
elmax = errorL[L.index(lmax)]
cluster      = [x/cmax for x in C]              # Normalize
lenght       = [x/lmax for x in L]
errorcluster = np.zeros(20)
errorlenght  = np.zeros(20)

for i in range(20):
    errorcluster[i] = np.sqrt((errorC[i]/cmax)**2+((errorC[i]*ecmax)/(cmax)**2)**2) # Error Propagation
    errorlenght[i]  = np.sqrt((errorL[i]/lmax)**2+((errorL[i]*elmax)/(lmax)**2)**2)

ax = plt.axes()
ax.set_xscale("log")
ax.errorbar(p,cluster,yerr=errorcluster, fmt = 'rx')
ax.errorbar(p,lenght,yerr=errorlenght, fmt = 'k.')
ax.legend(['Clustering Coefficient', 'Path Lenght'])
plt.title("Family of Randomly Rewired Graphs")
plt.xlabel("p")
plt.ylabel("Normalized C(p) and L(P)")
plt.show()
```
