```python
def dayOfProgrammer(year):
    # Write your code here
    psum, day, month = 0, 0, 0
    dpm = [31,0,31,30,31,30,31,31,30,31,30,31]
    if (year <=1917): #Julian
        if ((year%4)==0):
            dpm[1] = 29
        else:
            dpm[1] = 28
        
        
    elif (year == 1918): #Transition
        dpm[1] = 15 # In this transition year, february starts on 14th, but since it its not a leap
                    # year by neither Julian nor Gregorian calendar criteria, it has 28 days. Meaning
                    # the month has '15' days in total.
        
    elif (year >=1919): #Gregorian
        if ((year%400==0) or ((year%4==0) and (year%100!=0))):
            dpm[1] = 29
        else:
            dpm[1] = 28

    for i in range(12):
        for j in range(1,dpm[i]+1): # day 00 does not exist
            psum += 1
            if (psum==256):
                day = j
                month = i+1 # month 00 does not exist
    
    if (month<=9):
        month = str(''.join([str(0),str(month)]))
    
    dd, mm, yyyy = str(day), str(month), str(year)
    date = '.'.join([dd, mm, yyyy])
    return(date)
```
