```python
from sympy import *
x, y, z, a = symbols('x y z a')
init_printing(use_unicode=True)
velocidadpotencial = (x*(x**2+y**2+a**2))/(x**2+y**2)
potencial = (y*(x**2+y**2-a**2))/(x**2+y**2)
velocidadpotencial
```

```python
xcomp = diff(velocidadpotencial, x)
```

```python
ycomp = diff(velocidadpotencial, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
lim = 10
eppr = 18
x,y = np.meshgrid(np.linspace(-lim,lim,eppr),np.linspace(-lim,lim,eppr))
```

```python
a=7
i = (2*x**2/(x**2 + y**2) - 2*x**2*(a**2 + x**2 + y**2)/(x**2 + y**2)**2 + (a**2 + x**2 + y**2)/(x**2 + y**2))/np.sqrt((2*x*y/(x**2 + y**2) - 2*x*y*(a**2 + x**2 + y**2)/(x**2 + y**2)**2)**2 + (2*x**2/(x**2 + y**2) - 2*x**2*(a**2 + x**2 + y**2)/(x**2 + y**2)**2 + (a**2 + x**2 + y**2)/(x**2 + y**2))**2)
j = (2*x*y/(x**2 + y**2) - 2*x*y*(a**2 + x**2 + y**2)/(x**2 + y**2)**2)/np.sqrt((2*x*y/(x**2 + y**2) - 2*x*y*(a**2 + x**2 + y**2)/(x**2 + y**2)**2)**2 + (2*x**2/(x**2 + y**2) - 2*x**2*(a**2 + x**2 + y**2)/(x**2 + y**2)**2 + (a**2 + x**2 + y**2)/(x**2 + y**2))**2)

figure, axes = plt.subplots()
Drawing_colored_circle = plt.Circle((0, 0),a, color='black')
 
axes.set_aspect( 1 )
axes.add_artist( Drawing_colored_circle )
plt.title( 'Circle Obstacle' )
plt.quiver(x,y,i,j)
plt.show()
```
